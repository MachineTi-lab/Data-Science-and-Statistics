```python
import seaborn as sns
diamonds = sns.load_dataset("diamonds")
diamonds = diamonds.select_dtypes("float64" , "int64" ) 
df = diamonds.copy()
df.head()
```

```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
df = df.dropna() # fit_predict doesnt work so ı clean NAN data
```

```python
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1)
```

```python
clf.fit_predict(df) 
```

The fit_predict method didn’t work because the dataset contained NaN values. So, I cleaned the data before using fit_predict. Of course, this removed all rows with missing values, but alternatively, those gaps could have been filled with the mean values.

```python
df_scores = clf.negative_outlier_factor_
df_scores[0:10]
```
